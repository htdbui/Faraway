# 3. Factorial Experiments
- Suppose we have factors α, β, γ, etc., at levels $ l_α, l_β, l_γ, ... $ . A full factorial experiment includes at least one run for each combination of levels, totaling $ l_α l_β l_γ ,\cdots $, which can be very large. The largest model includes all possible interactions, from second-order or two-way to high-order involving several factors. Thus, full factorials are rarely used for more than three or four factors.
- Factorial designs have advantages. If no interactions are significant, we get multiple one-way experiments for the cost of one. Instead of replicating a two-factor experiment, we might investigate a third factor without replication.
- Analyzing full factorial experiments extends two-way ANOVA. Often, due to cost, there is no replication, so we assume higher-order interactions are zero to free degrees of freedom for testing lower-order effects. This assumption is reasonable since few phenomena require precise combinations of several factors.
- Fractional factorials use only a fraction of the runs in a full factorial experiment to save costs or due to limited experimental material. It is often sufficient to estimate lower order effects with just a fraction. For example, consider an experiment with seven factors, each at two levels:
| Effect | mean | main | 2-way | 3-way | 4 | 5 | 6 | 7 |
|-|-|-|-|-|-|-|-|-|
|# of parameters|1|7|21|35|35|21|7|1|
- If higher-order interactions are negligible, we don't need $2^7 = 128$ runs to estimate the remaining parameters. We could perform eight runs to estimate the seven main effects but not the interactions. Estimating all two-way interactions uniquely requires many runs, and designing good experiments is complex (see Hamada and Wu, 2000). A Latin square (Section 17.2) is another example of a fractional factorial. These experiments estimate many parameters with few data points, often leaving few degrees of freedom. We require small $\sigma^2$ to distinguish significant effects. Fractional factorials are popular in engineering and product design for screening many factors but are less useful in social sciences and medicine due to higher $\sigma^2$ and less control.

```python
lmod = smf.ols('y ~ h+d+l+b+j+f+n+a+i+e+m+c+k+g+o', speedo).fit()
lmod.sumary()
```

- There are no degrees of freedom as there are as many parameters as cases, preventing any usual tests. Understanding the coding is crucial, so examine the X-matrix:

```python
import patsy
dm = patsy.dmatrix('~ h+d+l+b+j+f+n+a+i+e+m+c+k+g+o', speedo)
np.asarray(dm)
```

- "+" is coded as zero and "-" as one due to their order in the ASCII alphabet.
- Without degrees of freedom, we cannot perform usual F-tests and need a different method to determine significance. Assuming no significant effects and normally distributed errors, the estimated effects are linear combinations of errors and hence normal. We make a normal quantile plot of the main effects, where outliers indicate significant effects:

```python
# Sort the indices of the parameters (excluding the intercept) in ascending order
ii = np.argsort(lmod.params[1:])

# Sort the coefficients (excluding the intercept) using the sorted indices
scoef = np.array(lmod.params[1:])[ii]

# Get the column names (excluding the last one) and sort them using the sorted indices
lcoef = (speedo.columns[:-1])[ii]

# Get the number of sorted coefficients
n = len(scoef)

# Calculate the normal quantiles for plotting
qq = stats.norm.ppf(np.arange(1, n+1) / (n+1))

# Create a new figure and axis for the plot
fig, ax = plt.subplots()

# Create a scatter plot of the normal quantiles versus the sorted coefficients
ax.scatter(qq, scoef, s=1)

# Annotate each point with the corresponding column name
for i in range(len(qq)):
    ax.annotate(lcoef[i], (qq[i], scoef[i]))
plt.xlabel("Normal Quantiles"); plt.ylabel("Sorted coefficients");
```

- Notice that "e" and possibly "g" are extreme. The negative "e" effect means the "+" level increases the response, which is undesirable for shrinkage, so we prefer the "-" level of "e". Similarly, the "+" level of "g" likely decreases shrinkage.
- A half-normal plot is better for detecting extreme points:

```python
# Get the number of sorted coefficients
n = len(scoef)

# Calculate the normal half quantiles for plotting
qq = stats.norm.ppf((n + np.arange(1, n+1)) / (2*n + 1))

# Get the absolute values of the coefficients (excluding the intercept)
acoef = np.abs(lmod.params[1:])

# Sort the indices of the absolute coefficients in ascending order
ii = np.argsort(acoef)

# Sort the absolute coefficients using the sorted indices
acoef = acoef.iloc[ii]

# Get the column names (excluding the last one) and sort them using the sorted indices
lcoef = (speedo.columns[:-1])[ii]

# Create a new figure and axis for the plot
fig, ax = plt.subplots()

# Create a scatter plot of the normal half quantiles versus the sorted absolute coefficients
ax.scatter(qq, acoef, s=1)

# Annotate each point with the corresponding column name
for i in range(len(qq)):
    ax.annotate(lcoef[i], (qq[i], acoef.iat[i]))
plt.xlabel("Normal Half Quantiles"); plt.ylabel("Sorted absolute coefficients");
```
