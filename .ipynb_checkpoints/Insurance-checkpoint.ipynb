# 1. Describe the data
- It has 47 rows.
- There are 8 variables:
    - race racial composition in percentage of minority. For example, if the value of the race variable is 30, it means that 30% of the population in that area is from racial minority groups.
    - fire fires per 100 housing units.
    - theft thefts per 1000 population.
    - age percentage of housing units built before 1939.
    - involact new FAIR (Fair Access to Insurance Requirements) plan policies and renewals per 100 housing units. Specifically, it counts the number of new policies and renewals issued per 100 housing units. The FAIR plan is designed to provide insurance to individuals who might otherwise be unable to obtain it due to factors like redlining, which is the discriminatory practice of denying services (such as insurance) to residents of certain areas based on their race or ethnicity. So, a higher involact value indicates more FAIR plan policies and renewals in that area, suggesting that more residents are relying on this plan for their insurance needs. In other words, a higher involact value reflects a greater dependence on the FAIR plan, which could be due to factors like economic challenges or discriminatory practices in the regular insurance market.
    - income median family income in thousands of dollars.
    - side north or south side of Chicago.
- Insurance redlining is the practice of denying insurance to specific people or areas. The term comes from drawing red lines on maps. While refusing auto insurance to frequent drunk drivers is generally accepted, other types of discrimination are not.
- In the late 1970s, the US Commission on Civil Rights investigated claims from Chicago community groups that insurance companies were redlining their neighborhoods. Due to a lack of data on individuals denied homeowners insurance, they recorded the number of FAIR plan policies issued and renewed by zip code from December 1977 to May 1978. The FAIR plan was a default policy for homeowners rejected by the voluntary market. They also collected data on other factors like fire and theft rates at the zip code level.

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
import faraway.utils
```

- We don’t know the races of those denied insurance, only the racial makeup of their zip codes. This is a key challenge to consider before starting the analysis.
- When data are collected at the group level, we might see a correlation between two variables. The ecological fallacy is assuming this correlation applies to individuals. For example, countries with higher dietary fat intake show higher breast cancer rates, but this doesn’t mean individuals with high fat intake are at greater risk. Observational data can be confounded, and aggregating data introduces bias.

```python
import faraway.datasets.eco
eco = faraway.datasets.eco.load()
eco.head()
```

```python
plt.scatter(eco.usborn, eco.income)
plt.ylabel("Income"); plt.xlabel("Fraction US born");
```

- We observe a negative correlation between 1998 per capita income and the proportion of U.S.-born residents in 1990 across the 50 states and D.C.
- A regression line can be fitted and displayed over an extended range.

```python
lmod = sm.OLS(eco.income, sm.add_constant(eco.usborn)).fit()
lmod.sumary()
```

```python
plt.scatter(eco.usborn, eco.income)
plt.ylabel("Income"); plt.xlabel("Fraction US born"); plt.xlim([0.6,1])
xr = np.array([0,1])
plt.plot(xr,lmod.params.iat[0] + xr*lmod.params.iat[1]);
```

- There is a statistically significant relationship between per capita income and the proportion of U.S.-born residents. Substituting values into the regression equation suggests naturalized citizens earn three times more than U.S.-born citizens. However, Census data shows U.S.-born citizens actually have a slightly higher average income than naturalized citizens. why?
- Ecological inference from aggregate data to individuals assumes income constancy, meaning native-born incomes don’t depend on their state’s native-born proportion (and similarly for naturalized citizens). This assumption is flawed because immigrants tend to move to wealthier states.
- This assumption is relevant to the Chicago insurance data analysis since we only have aggregate data. We must remember that results for aggregated data may not apply at the individual level.

# 4. Full Model and Diagnostics

```python
lmod = smf.ols('involact ~ race + fire + theft + age + np.log(income)', chredlin).fit()
lmod.sumary()
```

```python
# Check model assumptions
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
# Scatter plot of fitted values vs residuals
ax1.scatter(lmod.fittedvalues, lmod.resid)
ax1.set_ylabel("Residuals"); ax1.set_xlabel("Fitted values")
ax1.axhline(0, color='red', linestyle='--')
# Q-Q plot of residuals
sm.qqplot(lmod.resid, line="q", ax=ax2)
# "q": Quantile line
# "r": Regression line
# "s": Standardized line
# None: No line
fig.tight_layout()
```

- The diagonal streak in the residual-fitted plot is due to many zero response values. When y = 0, the residual $\hat{ε} = -y = -x^T\hat{β}$, hence the line . Ignoring this, there are no major issues.
- The Q-Q plot also looks good, which is reassuring given the response’s many zero values. We’ll rely on the central limit theorem, sample size, and lack of skewed residuals to trust the p-values.
- We try some partial residual plots with transformations:

```python
pr = lmod.resid + chredlin.race * lmod.params['race']
# Create a scatter plot of race vs partial residuals
plt.scatter(chredlin.race, pr)
plt.xlabel("Race")
plt.ylabel("Partial Residuals")

# Get the current x-axis limits
xr = np.array(plt.xlim())

# Plot the regression line for the race variable
# This line represents the effect of the race variable on the response
plt.plot(xr, xr * lmod.params['race'], color='red');
```

```python
pr = lmod.resid + chredlin.fire * lmod.params['fire']

# Create a scatter plot of fire vs partial residuals
plt.scatter(chredlin.fire, pr)
plt.xlabel("Fire")
plt.ylabel("Partial Residuals")

# Get the current x-axis limits
xr = np.array(plt.xlim())

# Plot the regression line for the fire variable
# This line represents the effect of the fire variable on the response
plt.plot(xr, xr * lmod.params['fire'], color='red');
```

- These plots show no need for transformation. Transforming the race variable would complicate interpretation, but luckily, it’s unnecessary. We checked other partial residual plots and tested polynomial predictors, finding no worthwhile transformations.
- We decided against transforming the response variable. The zero values limit options and complicate interpretation. While a square root transformation is possible, it offers minimal benefit and makes explanations harder.

# 5. Sensitivity Analysis

- How robust is our conclusion to the choice of covariates? In the full model, we used all four covariates, but we might question the sensitivity of our findings to this choice. For instance, should we adjust for the average income of the zip code? Similar concerns could apply to the other covariates.
- We can address these concerns by fitting models with different combinations of covariates. With four covariates, there are 16 possible combinations, making it practical to fit and examine all these models.

```python
import itertools

# List of indices
inds = [1, 2, 3, 4]

# Initialize an empty list to store combinations
clist = []

# Loop over the range from 0 to the length of inds + 1
for i in range(0, len(inds) + 1):
    # Generate all combinations of inds of length i
    # and extend clist with these combinations
    clist.extend(itertools.combinations(inds, i))

# clist now contains all possible combinations of the elements in inds
print(clist)
```

```python
chredlin.columns
```

```python
# Select specific columns from the chredlin dataset
X = chredlin.iloc[:, [1, 2, 3, 4, 5]].copy()

# Apply log transformation to the 'income' column
X.loc[:, 'income'] = np.log(chredlin['income'])

# Initialize lists to store the coefficients and p-values for the 'race' variable
betarace = []
pvals = []

# Loop over all combinations of the selected columns
for k in range(0, len(clist)):
    # Fit an OLS model with the 'involact' as the response variable
    # and the selected combination of predictors plus a constant
    lmod = sm.OLS(chredlin.involact, 
                  sm.add_constant(X.iloc[:, np.append(0, clist[k])])).fit()
    
    # Append the coefficient of the 'race' variable to betarace
    betarace.append(lmod.params.iat[1])
    
    # Append the p-value of the 'race' variable to pvals
    pvals.append(lmod.pvalues.iat[1])

# betarace now contains the coefficients of the 'race' variable for each model
# pvals now contains the p-values of the 'race' variable for each model
```

```python
# Initialize a list with the first variable 'race'
vlist = ['race']

# Array of variable names corresponding to the columns in X
varnames = np.array(['race', 'fire', 'theft', 'age', 'logincome'])

# Loop over all combinations of the selected columns, starting from the second combination
for k in range(1, len(clist)):
    # Create a string of variable names separated by '+'
    # np.append(0, clist[k]) ensures 'race' is always included
    vlist.append('+'.join(varnames[np.append(0, clist[k])]))

# vlist now contains strings representing different combinations of variables
print(vlist)
```

```python
pd.DataFrame({'beta':np.round(betarace, 4), 'pvals':np.round(pvals, 4)}, index=vlist)
```

- The output shows the $\hat{β}_1$ values and p-values for all 16 models. While $\hat{β}_1$ varies, with the highest value about double the lowest, the p-value never exceeds 5%. This means the effect’s significance is not sensitive to the choice of covariates, despite some uncertainty in its magnitude.
- If the outcome wasn’t clear and we found models where the predictor (race) wasn’t significant, the investigation would become more complex. We would need to carefully consider which covariates to adjust for. This discussion is beyond this book’s scope but shows why causal inference is challenging.
- We should check if our conclusions are affected by including or excluding a few cases. Influence diagnostics help with this. We start by computing the scaled change in coefficients when omitting a single case, focusing on whether this could make the race variable statistically significant.

```python
# Fit an OLS model with involact as the response variable and specified predictors
lmod = smf.ols('involact ~ race + fire + theft + age + np.log(income)', 
    chredlin).fit()

# Get influence diagnostics for the fitted model
diagv = lmod.get_influence()

# Find the minimum value of the DFBETAS for the 'race' variable
# DFBETAS measures the influence of each observation on the estimated coefficient of 'race'
min(diagv.dfbetas[:,1])
```

- The largest reduction in the t-statistic is about 0.4. Given the full data value of 3.82, this reduction is too small to alter our conclusions.
- We should also consider the influence on the adjustment covariates. We plot the leave-one-out differences in the scaled $\hat{β}$ for theft and fire.

```python
# Scatter plot of DFBETAS values for 'fire' and 'theft' variables
plt.scatter(diagv.dfbetas[:,2], diagv.dfbetas[:,3])
plt.xlabel("Change in Fire"); plt.ylabel("Change in Theft")
plt.axhline(0); plt.axvline(0);
```

```python
# Let’s also examine the standardized residuals and leverage
sm.graphics.influence_plot(lmod);
```

- The circle diameters represent the Cook Statistic size. Cases 5 and 23 stand out.
- We should examine other leave-one-out coefficient plots. Notably, there are no standardized residuals extreme enough to be outliers. Let's review the two cases.

```python
chredlin.iloc[[5, 23],:]
```

- These are high theft and fire zip codes. See what happens when we exclude these points in the model:

```python
ch45 = chredlin.drop(chredlin.index[[5,23]])
lmod45 = smf.ols('involact ~ race + fire + theft + age + np.log(income)', ch45).fit()
lmod45.sumary()
```

- The coefficient for race is lower than in the full data fit but still statistically significant. Additionally, theft and age are no longer significant.
- We've confirmed that our conclusions remain robust even if one or two cases are excluded. This is reassuring, as relying on a single case's accuracy would be concerning. If such influential cases were dropped, it would require strong justification and should be disclosed in the analysis.
- If we try hard to challenge our result, we can find a model where two cases have been excluded:

```python
lmodalt = smf.ols('involact ~ race + fire + np.log(income)', ch45).fit()
lmodalt.sumary()
```

- In this model, race is no longer significant. However, there’s no strong reason to favor this model over the many others we’ve considered.
- This highlights a common issue in regression modeling: data often doesn't clearly suggest a single model. Independent analysts can use similar methods differently, leading to varied model choices. Therefore, a good analyst thoroughly explores the data and considers multiple models.
- Confidence in conclusions increases if different models yield similar results. Our analysis shows this concern for alternatives, but human judgment is still crucial. An unscrupulous (dishonest) analyst might explore many models but only report the one that supports a specific conclusion.
- We don’t know the true model for this data, so our conclusions should reflect this uncertainty. Regression summaries give standard errors and p-values, but they don’t show model uncertainty, making us too confident in our results. To address this, we can split the data, use bootstrapping, or try model averaging.





# 6. Discussion
- The conclusion is unclear due to several reasons. The response isn’t a perfect measure of people being denied insurance and is an aggregate measure, which raises the issue of ecological correlations. We assumed the probability of a minority homeowner getting a FAIR plan, after adjusting for other factors, is constant across zip codes, which is unlikely. If this probability varies systematically, our conclusions may be incorrect. It would be better to get individual-level data.
- We found a statistically significant effect of race on the response, but this doesn’t mean it’s practically significant. The highest response value is only 2.2%, with most values much smaller. Our models predict about a 1% difference between 0% and 100% minority. So, while some people are affected, there may not be many. We need more data on predictors like insurance renewal rates to understand better. The p-value size doesn’t tell us much about the practical effect size.
- A hidden factor might actually be causing the observed relationship. Someone experienced in insurance might suggest one. This always makes our conclusions uncertain.
- Another issue is how much to combine the data. For example, fit separate models for the two halves of the city.

```python
# South
lmods = smf.ols('involact ~ race + fire + theft + age', chredlin.loc[chredlin.side == 's',:]).fit()
lmods.sumary()
```

```python
# North
lmodn = smf.ols('involact ~ race + fire + theft + age', chredlin.loc[chredlin.side == 'n',:]).fit()
lmodn.sumary()
```

- Race is significant in the north but not in the south. Dividing data into smaller subsets can dilute any predictor’s significance. However, it’s also important not to combine all data without considering if it’s reasonable. This judgment can be contentious in legal cases.
- Explaining this in court is tough. Scientific studies often have uncertainties, especially if the topic isn't controversial. In legal battles, unclear statistical evidence is hard to present. It's also difficult to explain to people without math training.
- After all this analysis, the reader might feel a bit dissatisfied. It seems we can't reach any definite conclusions, and everything is full of "ifs" and "buts." Winston Churchill once said that democracy is the worst form of government except for all the others that have been tried. We might say the same about statistics in dealing with uncertainty. It's not perfect, but the alternatives are worse.
