# 1. Describe the data
- It has 88 rows.
- There are 3 variables:
    - NOx: concentration of nitrogen oxides (NOx) emissions from the engine. The unit is typically parts per million (ppm).
    - C: concentration of ethanol in the fuel mixture. The unit is typically a percentage (%).
    - E: equivalence ratio, which is the ratio of the actual fuel-to-air ratio to the stoichiometric fuel-to-air ratio. It is a dimensionless quantity.
- Ethanol fuel was burned in a single-cylinder engine. The NOx emissions varied based on the engineâ€™s compression ratio and the equivalence ratio (E).

# 3. Polynomial
- To make the $X\beta$ part of the model more flexible, we can add polynomial terms. For a single predictor, the model looks like this: $y = \beta_0 + \beta_1x + \ldots + \beta_dx^d + \varepsilon$. This helps the model capture more complex patterns in the data. While polynomials might not perfectly represent the real relationship, they can highlight certain features. For instance, a quadratic term $x^2$ can indicate an optimal value for the predictor, similar to finding the best temperature for baking bread. If we believe a predictor behaves this way, adding a quadratic term is sensible.
- Two common methods to choose (d) are: adding terms until the new term is not statistically significant, or starting with a high-degree polynomial and removing non-significant terms, beginning with the highest power.
- The relationship between C and NOx shows a clear peak, making a linear term insufficient. The simplest model includes a quadratic term for E:

```python
lmod2 = smf.ols('NOx ~ E + I(E**2) + C', ethanol).fit()
lmod2.sumary()
```

```python
# Finding the value of E that maximizes NOx output
-lmod2.params.iat[1]/(2*lmod2.params.iat[2])
```

```python
# Using numerical optimization to find the maximum
from scipy.optimize import minimize_scalar
# minimize_scalar finds the minimum of a scalar function
f2 = np.poly1d(lmod2.params.iloc[[2,1,0]])
# np.poly1d creates a polynomial function
# lmod2.params[[2,1,0]] selects the coefficients in reverse order
result = minimize_scalar(-f2)
# minimize_scalar(-f2) finds x that minimizes negated polynomial function, which
# is equivalent to finding x that maximizes original polynomial function f2
result.x
```

```python
# Checking the model fits the data well with a residual check such as: 
plt.scatter(ethanol.E, lmod2.resid);
```

- The plot reveals some remaining structure, suggesting that a more flexible function than a quadratic is required. A quartic function might be preferable over a cubic to achieve the expected shape for predicting NOx.

```python
lmod4 = smf.ols('NOx ~ E + I(E**2) +I(E**3) + I(E**4) + C',
    ethanol).fit()
lmod4.sumary()
```

- The residual checks are much more satisfactory, and the R-squared indicates a better fit. However, there are drawbacks. The estimated coefficients are very large, often due to strong collinearities between polynomial terms. This can negatively impact the stability of the estimates and predictions.
- In this data set, we see that the predicted response cannot be negative. To ensure this, we can use a logged response.

```python
lmodl = smf.ols('np.log(NOx) ~ E + I(E**2) + C', ethanol).fit()
lmodl.sumary()
```

- This fit is superior to the unlogged quadratic model.

```python
# Defining functions associated with the quartic and logged quadratic models
f4 = np.poly1d(lmod4.params.iloc[[4,3,2,1,0]]) # selects coefficients in reverse order
fl = np.poly1d(lmodl.params.iloc[[2,1,0]]) # selects coefficients in reverse order
# Plot all three fits
# Scatter plot of E vs NOx
plt.scatter(ethanol.E, ethanol.NOx, label=None)
# Generate a range of values for E
z = np.linspace(0.4, 1.4)
# Calculate the mean of C
meanC = np.mean(ethanol.C)
# Plot the quadratic model
plt.plot(z, f2(z) + meanC * lmod2.params['C'], 'k-', label='quadratic')
# Plot the quartic model
plt.plot(z, f4(z) + meanC * lmod4.params['C'], 'k--', label='quartic')
# Plot the log-quadratic model
plt.plot(z, np.exp(fl(z) + meanC * lmodl.params['C']), 'k:', label='log-quadratic')
plt.legend();
```

- I computed predictions over a slightly wider range of E values than observed in the data, using the mean value of C.
- All three models fit the data well but become unstable when extrapolated. Polynomial functions, especially higher-order ones, tend to produce unrealistic results as values become extreme.
- In this data, we used a logged response to maintain sensible behavior outside the main data range, but this approach isnâ€™t always feasible.

```python
# Finding the optimizing value of E for the quartic model
minimize_scalar(-f4).x
```

- This causes many warnings and an unrealistically high value.
- Figure above shows we need to limit the maximization range to capture the central peak.

```python
minimize_scalar(-f4,bounds=(0.6,1.2), method="bounded").x
```

```python
# Finding the optimizing value of E for the logged quadratic model
minimize_scalar(-fl).x
```

- Despite their differences, the three models have similar maximizing values. This is expected due to the dataâ€™s symmetry.
- Using polynomials higher than quadratic is problematic due to collinearity and unstable extrapolation. The coefficients lack useful interpretation. For a more flexible fit, consider piecewise linear or spline methods.
- If prefer using polynomials, consider Legendre polynomials. These are polynomials $ P_n(x) $ of degree $ n $ defined on $[-1, 1]$ which have the orthogonal property:
    - $\int_{-1}^{1} P_m(x)P_n(x)dx = 0$
    - If we normalize by requiring that $ \int_{-1}^{1} P_n(x)^2 \, dx = 1 $, this is sufficient to uniquely define them. The first few ones are $ 1 $, $ x $, $ \frac{3x^2 - 1}{2} $, $ \frac{5x^3 - 3x}{2} $.
- We use Legendre polynomials up the fourth degree:

```python
# Step 1: Scaling 'E' variable
cmax = max(ethanol.E)  # Maximum value of 'E'
cmin = min(ethanol.E)  # Minimum value of 'E'
# Linear transformation parameters
a = 2 / (cmax - cmin)  # Scaling factor to normalize 'E' to the range [-1, 1]
b = -(cmax + cmin) / (cmax - cmin)  # Translation to shift the scaled values
# Applying the scaling transformation
sC = ethanol.E * a + b  # sC is the transformed 'E' scaled between -1 and 1
# Step 2: Create design matrix using Legendre polynomials
# Generating a matrix of Legendre polynomial values up to degree 4 (for the scaled variable)
X = np.polynomial.legendre.legvander(sC, 4)
# Step 3: Add categorical variable 'C' to the design matrix
Ccol = np.array(ethanol['C']).reshape(-1, 1)  # Reshape 'C' to a column vector
# Concatenate Legendre polynomial matrix with 'C' column
X = np.concatenate((X, Ccol), axis=1)
# Step 4: Fit a linear model using Ordinary Least Squares (OLS)
lmodleg = sm.OLS(ethanol.NOx, X).fit()
lmodleg.sumary()
```

- First, scale the predictor ( E ) to ([-1, 1]). Use the legvander function to create the design matrix and append the column for ( C ).
- The fitted values match the quartic model, but the coefficients are stable.

```python
# Computing the correlations between the Legendre polynomial terms
np.corrcoef(X[:, 1:5].T).round(3) 
```

- These correlations are generally small.
- If predictor values were evenly spaced, correlations would be nearly zero due to the orthogonality of Legendre polynomials.
- Zero correlation is desirable for choosing the polynomial degree, as coefficients remain stable when terms are added or removed, allowing for a single model fit.
- Creating an orthogonal design with unevenly spaced predictors is possible but often not worth the effort. Orthogonal polynomials were useful when computing was costly, but they require extra scaling work. Today, the benefits donâ€™t justify the costs.
- Eliminating lower-order terms before higher-order ones, even if not statistically significant, is usually unwise due to the hierarchy principle. Removing an input typically simplifies predictions, but here, the input is still needed. Additionally, a scale change affects the t-statistics of all but the highest-order term, making conclusions sensitive to inconsequential scale changes. For example, transforming E by subtracting 0.9 and refitting the quadratic model illustrates this:

```python
ethanol['Ec'] = ethanol.E - 0.9
lmodc = smf.ols('NOx ~ Ec + I(Ec**2) + C', ethanol).fit()
lmodc.sumary()
```

- The quadratic term remains unchanged from the uncentered model, but the linear term is now insignificant. Thereâ€™s no reason to remove the linear term here but not in the previous model.
- Removing lower-order terms has specific implications: setting the intercept to zero means the regression passes through the origin, while setting the linear term to zero means the response is optimized at a predictor value of zero.
- You can also define polynomials in more than one variable. These are sometimes 
called response surface models. A second degree model would be: $ y = Î²_0 + Î²_1 x_1 + Î²_2 x_2 + Î²_{11} x_1^2 + Î²_{22} x_2^2 + Î²_{12} x_1 x_2 $.
- We choose a second-order model.

```python
from sklearn.preprocessing import PolynomialFeatures
# Step 1: Select the relevant columns
# We are using 'C' (categorical) and 'E' (continuous) as our predictor variables
X = ethanol.loc[:, ['C', 'E']]
# Step 2: Create polynomial features up to degree 2
poly = PolynomialFeatures(2)  # Degree 2 polynomial features (including interactions)
Xp = poly.fit_transform(X)  # Transform the data
# Step 3: Convert the transformed features into a DataFrame for better readability
Xp = pd.DataFrame(Xp, columns=['Intercept', 'C', 'E', 'C2', 'E2', 'CE'])
# Step 4: Fit an OLS model using the polynomial features
lmod = sm.OLS(ethanol.NOx, Xp).fit()  # Fitting the OLS model
lmod.sumary()
```

- All terms are strongly significant except the quadratic term of C. The fitted surface shows little curvature in the C direction.
- The CE interaction term represents the surfaceâ€™s rotation relative to the axes. Without it, we could optimize one variable independently, as in the previous analysis. This model indicates we cannot maximize E without considering C.
- Although the quadratic term in C is not significant, removing it offers little benefit. We respect the hierarchy principle since the second-order terms E2 and CE are significant.
- This model shows the simplest way to allow interaction between two quantitative predictors. Including the interaction term CE means including all three second-order terms.

```python
# Step 1: Define the grid size for 'C' and 'E'
ngrid = 11  # Number of grid points
# Step 2: Generate equally spaced values for 'C' and 'E'
# Cx is the grid of values for 'C' (categorical), evenly spaced
Cx = np.linspace(min(ethanol.C), max(ethanol.C), ngrid)
# Ey is the grid of values for 'E' (continuous), evenly spaced
Ey = np.linspace(min(ethanol.E), max(ethanol.E), ngrid)
# Step 3: Create a grid of values for 'C' and 'E'
X, Y = np.meshgrid(Cx, Ey)  # X and Y are 2D grids for the 'C' and 'E' values
# Step 4: Combine the 'C' and 'E' grid values into pairs
XYpairs = np.dstack([X, Y]).reshape(-1, 2)  # Create pairs of ('C', 'E')
# Step 5: Generate polynomial features from the grid points
poly = PolynomialFeatures(2)  # Using the same degree-2 polynomial model as before
Xp = poly.fit_transform(XYpairs)  # Generate polynomial features for the grid points
# Step 6: Predict the NOx values using the model's coefficients
pv = np.dot(Xp, lmod.params)  # Predicted NOx values using dot product of features and model parameters
# Step 7: Reshape the predicted values into a grid
Z = np.reshape(pv, (ngrid, ngrid))  # Reshape the predictions into a 2D grid corresponding to 'C' and 'E'
from mpl_toolkits.mplot3d import Axes3D
# Step 8: Plot the 3D wireframe plot
fig = plt.figure()  # Create a new figure
ax = fig.add_subplot(111, projection='3d')  # Add a 3D subplot
ax.plot_wireframe(X, Y, Z)  # Plot the wireframe for 'C', 'E', and predicted 'NOx'
ax.set_xlabel("C"); ax.set_ylabel("E"); ax.set_zlabel("NOx");
```

# 4. Additive Models
- Finding the right transformations for predictors can be challenging, especially when dealing with multiple predictors. Adjusting the transformation for one predictor might affect the optimal transformation for another. Luckily, thereâ€™s a method to select transformations for all predictors at once. This method involves using an additive model, which takes the form:
$$ y = Î± + f_1 (X_1) + f_2 (X_2) + \cdots + f_p (X_p) + Îµ $$
- The linear terms like $Î²_i X_i$ have been replaced with more flexible functions $f_i(X_i)$. Models that can handle non-normal responses are known as Generalized Additive Models (GAM). The statsmodels package includes some capabilities for fitting GAMs, although we specifically need the generalized component here.
- We need to define the general functional forms for $f_1$ and $f_2$. B-splines are commonly used, with cubic (degree=3) being the standard choice. We also need to determine the number of splines to use in each dimension. For a relatively small dataset like this one, using a smaller number, such as 4, is appropriate. Larger datasets would benefit from the increased flexibility provided by using more splines. Before fitting the model, we must set up the B-spline basis.

```python
from statsmodels.gam.api import GLMGam, BSplines
# Step 2: Define the predictor matrix 'xmat'
# 'xmat' consists of two columns: 'C' and 'E' from the ethanol dataset.
xmat = ethanol[['C', 'E']]
# Step 3: Define B-splines for the predictor variables
# BSplines is used to create a set of basis functions for the 'C' and 'E' variables.
# We specify:
# df=[4, 4]: The number of degrees of freedom for the splines (4 for both 'C' and 'E')
# degree=[3, 3]: The degree of the B-spline (3, cubic spline for both 'C' and 'E')
bs = BSplines(xmat, df=[4, 4], degree=[3, 3])
# Step 4: Fit a Generalized Additive Model (GAM)
# GLMGam.from_formula is used to specify and fit the GAM model.
# We use 'NOx ~ C + E' as the formula, meaning we model 'NOx' as a function of 'C' and 'E'.
# 'smoother=bs' passes the B-splines defined earlier as the smoothing functions for the predictors.
gamod = GLMGam.from_formula('NOx ~ C + E', data=ethanol, smoother=bs).fit()
gamod.summary()
```

- Examining the fitted coefficients from this model can be challenging to interpret. Instead, itâ€™s more useful to plot the predicted response as a function of one variable while keeping other variables constant, as shown in the pair of plots below. These plots also display partial residuals and 95% confidence bands.

```python
# Step 1: Plot the partial dependence for the first variable (C)
# 'plot_partial(0)' plots the partial dependence of the first predictor (C) in the model
# 'cpr=True' indicates that we want to plot the component plus residual (CPR) plot, 
# which shows how the predictor (C) affects the response (NOx) while accounting for the residuals.
fig = gamod.plot_partial(0, cpr=True)
```

```python
# Step 2: Plot the partial dependence for the second variable (E)
# Similarly, 'plot_partial(1)' plots the partial dependence of the second predictor (E) in the model
# 'cpr=True' shows the component plus residual plot for the predictor (E).
fig = gamod.plot_partial(1, cpr=True)
```

1. **Partial Dependence Plot**:
   - A partial dependence plot shows the relationship between a particular predictor (C or E) and the response (NOx),
     while averaging out the effects of the other predictors.
   - The plot helps visualize the isolated effect of each predictor on the response, making it easier to understand
     how the predictor impacts the outcome.

2. **Component Plus Residual (CPR) Plot**:
   - The 'cpr=True' argument indicates that we are plotting the component plus residual plot.
   - This plot shows the fitted component of the predictor plus the residuals. The residuals are added to the plot
     to reveal any patterns that are not captured by the smooth function.

3. **`plot_partial(0)` and `plot_partial(1)`**:
   - `plot_partial(0)` refers to the first predictor in the model, which in this case is 'C'.
   - `plot_partial(1)` refers to the second predictor in the model, which is 'E'.

4. **Result**:
   - You will get two plots: one for 'C' and one for 'E'. Each plot will display:
       - The effect of the predictor on the response (NOx).
       - The residuals, which show how much variation remains after accounting for the fitted model.

5. **Visualization**:
   - These plots are useful for diagnosing the model. They help determine whether the fitted smooth relationship
     between the predictor and the response is adequate or whether there might be additional structure in the data
     that the model is missing.

- We observe a roughly linear relationship for variable C and a quadratic one for variable E, which aligns with our previous analyses.
- The true value of the GAM analysis, however, lies in its ability to reveal unexpected patterns, especially when dealing with more than two predictors.
- The GAM approach can serve as a comprehensive inference system or as an exploratory tool to find suitable transformations for predictors in linear models. Currently, due to the limitations of Python software, only the exploratory approach is feasible.
- Itâ€™s important to carefully select the amount of smoothing in a GAM. In our example, the default smoothing choice works well. The GAM functions in the statsmodels package can automatically determine the appropriate amount of smoothing based on various criteria.
