# 1. Describe the data
- It has 9 rows.
- There are 3 variables:
    - Strength: This is the measured strength of the composite material (in Mega Pascals MPa).
    - Laser: This indicates the power of the laser used, with values like 40W, 50W, and 60W.
    - Tape: This describes the speed of the tape, with categories such as slow, medium, and fast. Tape refers to the speed at which the tape moves during the process of making the composite material.
- Mazumdar and Hoa (1995) conducted an experiment testing the strength of a thermoplastic composite based on laser power and tape speed.
- The composite's strength was tested under varying laser power and tape speed. Laser power affects the bonding process, while tape speed influences processing time. The experiment aimed to find the optimal conditions for the strongest thermoplastic composite.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
from scipy import stats
import faraway.utils
```

```python
import faraway.datasets.composite
composite = faraway.datasets.composite.load()
composite
```

```python
# x: 'laser' indicates the x-axis variable, which represents the power of the laser
# y: 'strength' indicates the y-axis variable, which represents the strength of the composite
# hue: 'tape' groups the data points by the 'tape' variable, differentiating them by color
# data: 'composite' is the DataFrame containing the data to be plotted
# kind: 'point' specifies the type of plot to be a point plot
sns.catplot(x="laser", y="strength", hue="tape", data=composite, kind="point");
```

```python
# Create a new column 'Nlaser' in the composite DataFrame by tiling the array [40, 50, 60] three times
composite['Nlaser'] = np.tile([40,50,60], 3)

# Get unique values from the 'tape' column in the composite DataFrame and store them in faclevels
faclevels = np.unique(composite.tape)

# Define a list of line styles for plotting
lineseq = ['-', '--', ":"]

# Loop through each unique value in faclevels
for i in np.arange(len(faclevels)):
    # Create a boolean mask where 'tape' equals the current faclevel
    j = (composite.tape == faclevels[i])
    # Plot the 'Nlaser' and 'strength' values for the current faclevel with the corresponding line style and label
    plt.plot(composite.Nlaser[j], composite.strength[j], lineseq[i], label=faclevels[i])

plt.legend(); plt.xlabel("Laser"); plt.ylabel("Strength")
plt.xticks([40,50,60], ["40W", "50W", "60W"]);
```

```python
sns.catplot(x="tape", y="strength", hue="laser", 
    data=composite, kind="point");
```

```python
composite['Ntape'] = np.repeat([6.42,13,27], 3)
faclevels = np.unique(composite.laser)
for i in np.arange(len(faclevels)):
    j = (composite.laser == faclevels[i])
    plt.plot(composite.Ntape[j], composite.strength[j],
             lineseq[i],label=faclevels[i])
plt.legend(); plt.xlabel("Tape"); plt.ylabel("Strength")
plt.xticks([6.42,13,27],["slow","medium","fast"]);
```

- The first plot shows that medium tape speed maximizes strength, while the second plot indicates that the 60W laser setting is best. Parallel lines in each plot suggest additive effects with minimal interaction, despite some noise. In this example, the lines appear roughly parallel.

# 3. Models with Several Factors
- We learn to analyze data with multiple categories from observational studies or factorial design experiments. A full factorial design includes all possible category combinations at least once. Repeated observations for the same combination are called replicates.
## 3.1 Two Factors with No Replication
- A general model for this type of data: $y_{ij}=\mu+\alpha_i+\beta_j+\left(\alpha\beta\right)_{ij}+\varepsilon_{ij}$
- Where i = 1, …, I and j = 1, …, J. αβ is interactions between α and β.

```python
lmod = smf.ols("strength ~ tape * laser", composite).fit()
lmod.sumary()
```

- The model has a perfect fit with parameter estimates equal to the number of cases. However, no further inference is possible, offering little value from this model.
- We can't directly check for interaction by adding a term to the model, but Tukey's nonadditivity test offers an alternative method to investigate interaction. The model: $y_{ij}=\mu+\alpha_i+\beta_j+\phi\alpha_i\beta_j+\varepsilon_{ij}$
    - We fit the model and test if $\phi=0$.
    - This is a nonlinear model because it has $\phi\alpha_i\beta_j$, which is difficult to fit in general. We address this by fitting the model in stages: first, we fit the model without interaction and extract the main effects estimates.

```python
lmod = smf.ols("strength ~ tape + laser", composite).fit()
pd.DataFrame(lmod.params.index).T
```

- Since the tape effects are not in the expected order, we must construct them carefully (same for laser effects):

```python
# store slow tape and then medium tape
tapecoefs = np.repeat([lmod.params.iat[2], lmod.params.iat[1], 0], 3)
# store 50W laser and then 60W laser
lasercoefs = np.tile(np.append(0, lmod.params[3:5]), 3)
```

```python
# update model with new predictor formed from products of main effects:
# Create a new column 'crossp' in the composite DataFrame by multiplying tapecoefs with lasercoefs
composite['crossp'] = tapecoefs * lasercoefs

# Fit an Ordinary Least Squares (OLS) regression model with strength as the dependent variable
# and tape, laser, and crossp as the independent variables
tmod = smf.ols("strength ~ tape + laser + crossp", composite).fit()
tmod.sumary()
```

- The p-value of 0.5503 indicates a nonsignificant interaction, suggesting $(\alpha\beta)_{ij} = 0$. The test assumes a multiplicative interaction effect, which may not be accurate. Judging whether lines on a plot are parallel, though subjective, may be more reliable for checking interaction here.
- With interactions addressed, we can check the significance of the main effects:

```python
sm.stats.anova_lm(tmod)
```

- We see that both factors are significant.
- The treatment coding doesn't utilize the ordered nature of both factors. Nominal factors lack an order, while ordinal factors have a natural order. We can declare both as ordered factors and refit:

```python
# Define an ordered categorical type with categories 'slow', 'medium', and 'fast'
cat_type = pd.api.types.CategoricalDtype(
    categories=['slow', 'medium', 'fast'], ordered=True)

# Convert the 'tape' column in the composite DataFrame to this ordered categorical type
composite['tape'] = composite.tape.astype(cat_type)

# Import the Poly contrast from patsy for polynomial coding of categorical variables
from patsy.contrasts import Poly

# Fit an Ordinary Least Squares (OLS) regression model with 'strength' as the dependent variable
# and 'tape' and 'laser' as independent variables, using polynomial contrasts for both factors
lmod = smf.ols("strength ~ C(tape, Poly) + C(laser, Poly)", composite).fit()
lmod.sumary()
```

- Instead of reference level coding, we use linear and quadratic terms for each factor. The coding is:

```python
from patsy import dmatrix
# Create a design matrix using polynomial contrasts for 'tape' and 'laser' factors in the composite DataFrame
dm = dmatrix('~ C(tape, Poly) + C(laser, Poly)', composite)
# Convert the design matrix to a NumPy array and round the values to 2 decimal places
np.asarray(dm).round(2)
```

- The linear term, proportional to (-1, 0, 1), represents a linear trend, while the quadratic term, proportional to (1, -2, 1), represents a quadratic trend. The quadratic term for laser power is not significant, but there is a quadratic effect for tape speed. A drawback of factor models is the difficulty in extrapolating to new conditions. The ordered factors suggest using numerical predictors corresponding to the level values.

```python
# Fit an Ordinary Least Squares (OLS) regression model with 'strength' as the dependent variable
# Using the natural logarithm of 'Ntape' and its square, along with 'Nlaser', as independent variables
lmodn = smf.ols(
   "strength ~ np.log(Ntape) + I(np.log(Ntape)**2) + Nlaser", 
   composite).fit()
lmodn.sumary()
```

- Using the log of tape speed results in roughly evenly spaced levels. This model fits as well as the two-factor model but allows predictions for tape speed and laser power values not used in the experiment. The earlier factor analysis helped discover this model.
