# 1. Describe the data
- It has 48 rows.
- There are 3 variables:
    - psize: Particle size
    - operator: Operator number 1, 2 or 3
    - resin: Resin railcar 1-8
- 3 operators used 8 resin railcars (devices) to produce PVC in an experiment to study factors affecting its production. Each of the 24 combinations produced two samples, measuring the particle size of the product. This experiment is detailed in Morris and Watson (1998).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
from scipy import stats
import faraway.utils
```

```python
import faraway.datasets.pvc
pvc = faraway.datasets.pvc.load()
pvc.head()
```

```python
# Create a point plot to visualize the particle size ('psize') of PVC produced by 8 resin railcars ('resin'),
# with points colored by the operator ('operator'). No error bars are included in the plot.
sns.catplot(x='resin', y='psize', hue='operator', data=pvc, kind="point", errorbar=None);
```

```python
# Group the PVC dataset by 'operator' and 'resin', then calculate the mean of 'psize' for each group,
# and reset the index to get a flat DataFrame
pvcm = pvc.groupby(['operator', 'resin'])[['psize']].mean().reset_index()

# Get the unique levels of the 'operator' factor
faclevels = np.unique(pvcm.operator)

# Define a sequence of line styles to use for plotting
lineseq = ['-', '--', ':']

# Loop through each unique operator
for i in np.arange(len(faclevels)):
    # Create a boolean mask for the current operator
    j = (pvcm.operator == faclevels[i])    
    # Plot particle size ('psize') against resin for the current operator using the specified line style
    plt.plot(pvcm.resin[j], pvcm.psize[j], lineseq[i], label=faclevels[i])
plt.legend(title="Operator"); plt.xlabel("resin"); plt.ylabel("Particle Size");
```

```python
# Create a point plot to visualize the particle size ('psize') of PVC produced by 3 operators ('operator'),
# with points colored by the resin type ('resin'). The confidence interval for the error bars is specified by 'errobar'.
# The points are scaled by a factor of 0.5.
sns.catplot(x='operator', y='psize', hue='resin', data=pvc, errorbar=None, markersize=3, kind="point");
```

```python
# Get the unique levels of the 'resin' factor
faclevels = np.unique(pvcm.resin)

# Loop through each unique resin
for i in np.arange(len(faclevels)):
    # Create a boolean mask for the current resin
    j = (pvcm.resin == faclevels[i])
    
    # Plot particle size ('psize') against operator for the current resin
    plt.plot(pvcm.operator[j], pvcm.psize[j])
    
    # Annotate the plot with the index of the resin
    plt.annotate(str(i+1), [0.95, np.ravel(pvcm.psize[j])[0]])
plt.xlabel("Operator"); plt.xticks([1, 2, 3]); plt.ylabel("Particle Size");
# The eight resins are hard to distinguish with linetypes, 
# so we labeled them with numbers on the left.
```

- In the first plot, we have eight levels of resin, each with averaged responses from three operators, distinguished by different line types. The variances for each resin level appear equal, and the parallel lines suggest minimal interaction.
- In the second plot, lines connect the mean values for each resin level. Particle sizes are largest for the first operator and smallest for the third. The response variance is consistent across operators. The eight levels are labeled for clarity, and the parallel lines indicate little interaction between the two factors.
- Interaction plots can be challenging due to random variation, making it hard to distinguish true interaction from noise. Fortunately, with replication, we can directly test for interaction effects. We fit the full model and check the significance of the factors. Resin and operator must be converted to categorical variables to avoid fitting them as numeric predictors.

```python
pvc['operator'] = pvc['operator'].astype('category')
pvc['resin'] = pvc['resin'].astype('category')
```

# 3. Two Factors with Replication
- Consider the case when each combination of factor levels has the same number of observations greater than one. This layout results in an orthogonal design matrix. With replication, we can fit and test the full model:
$$ y_{ijk}=\mu+\alpha_i+\beta_j+\left(\alpha\beta\right)_{ij}+\varepsilon_{ijk} $$
- The interaction effect is tested by fitting a model without the $\left(\alpha\beta\right)_{ij}$ term and performing an F-test. If significant, we do not test the main effects even if they are not significant.

```python
# Fit an OLS model with particle size (psize) as the response variable
# and the interaction between operator and resin as predictor variables.
lmod = smf.ols('psize ~ operator*resin', pvc).fit()

# Perform ANOVA on the fitted model and round the results to 3 decimal places.
sm.stats.anova_lm(lmod).round(3)
```

- The interaction effect is not significant, allowing us to meaningfully investigate the main effects, both of which are significant. One may be tempted to remove the interaction and retest the main effects:

```python
lmod2 = smf.ols('psize ~ operator+resin', pvc).fit()
sm.stats.anova_lm(lmod2).round(3)
```

- The conclusions remain the same, but results differ slightly due to the $\hat{\sigma}^2$ used in the F-test denominator (1.478 in the first table and 1.311 in the second). The first method achieves greater power and is preferred.

# 4. Diagnostics

```python
sm.qqplot(lmod.resid, line="q");
```

```python
sns.residplot(x=lmod.fittedvalues,y=lmod.resid)
plt.xlabel("Fitted values"); plt.ylabel("Residuals");
```

```python
# Create a swarm plot using seaborn to visualize the residuals of the fitted model (lmod)
# with 'operator' on the x-axis and the residuals on the y-axis.
sns.swarmplot(x=pvc['operator'], y=lmod.resid)
plt.axhline(0); plt.xlabel("Operator"); plt.ylabel("Residuals");
```

- A swarm plot is a scatter plot that adjusts points on the categorical axis to prevent overlap, aiding in visualizing a variable's distribution across categories.
- The Q-Q plot suggests long-tailed errors. The residuals vs. fitted plot shows some symmetry. For each factor combination, the mean of the two replicates is the fitted value, and the two residuals are mirror images. Distant pairs mainly derive from operator 3, as seen in the last plot.
- Although the plot is convincing, we can test for variance differences. For each combination of operator and resin, compute the two residuals, one positive and one negative due to symmetry. Use $\sqrt{|\varepsilon|}$ as the new response to measure the spread, which reduces skewness. Only one observation per pair is needed since they are identical.

```python
# Fit an OLS model with particle size (psize) as the response variable
# and the interaction between operator and resin as predictor variables.
lmod = smf.ols('psize ~ operator*resin', pvc).fit()

# Create an array of indices for every second observation (0, 2, 4, ..., 46)
ii = np.arange(0, 48, 2)

# Select every second row from the original dataframe pvc and create a copy
pvce = pvc.iloc[ii, :].copy()

# Calculate the square root of the absolute residuals from the fitted model (lmod) for every second observation,
# and add this as a new column 'res' to the copied dataframe pvce
pvce['res'] = np.sqrt(abs(lmod.resid.loc[ii + 1]))

# Fit another OLS model with the new response variable 'res' and both operator and resin as predictors
vmod = smf.ols('res ~ operator+resin', pvce).fit()

# Perform ANOVA (Analysis of Variance) on the fitted model (vmod) and round the results to 3 decimal places
sm.stats.anova_lm(vmod).round(3)
```

- This confirms a significant difference in variation for the operators and no strong evidence against constant variation among the resin cars. Consistency in manufacturing is crucial, as high variance can harm production quality even if the mean is adjusted. Let's examine the main effects, noting that the lack of constant variance makes comparisons less optimal.

```python
lmod = smf.ols('psize ~ operator+resin', pvc).fit()
lmod.sumary()
```

- If we value small particle size, operator 3 produces the best results, and resin car 4 is preferred. However, if we cannot control a factor (e.g., disposable resin cars), identifying the best is pointless. Including the factor in the model still helps estimate operator effects more precisely.
- Pairwise confidence intervals for treatment factors can be constructed using the Tukey method, starting with the interval width for operator differences.

```python
# Import the get_tukeyQcrit function from the statsmodels library
from statsmodels.sandbox.stats.multicomp import get_tukeyQcrit

# Compute the Tukey Q critical value for 3 groups and 38 degrees of freedom
# This value is used to construct Tukey's Honest Significant Difference (HSD) intervals
tukey_q_crit = get_tukeyQcrit(3, 38)

# Compute the width of the confidence interval for the first coefficient in the model (lmod)
# lmod.bse[1] is the standard error of the first coefficient
# The interval width is scaled by the Tukey Q critical value and divided by the square root of 2
tukey_q_crit * lmod.bse.iat[1] / np.sqrt(2)
```

```python
# Compute all the pairwise differences among operators
p = np.append(0, lmod.params[1:3])
np.add.outer(p,-p)
```

- Operators 1 and 2 are not significantly different, but operator 3 is different from both. More significant differences exist among the resin cars, though this may not be of specific interest.
- The analysis is appropriate for specific operators and resin cars, treated as fixed effects. If operators and resin cars were randomly selected, they should be analyzed as random effects, requiring a different analysis. The variation between resin cars is greater than between operators.
- Observations in each cell must be genuine replications; otherwise, the analysis needs adjustment. Repeat measurements are not independent and can be handled with repeated measures models, requiring careful separation of measurements for each operator and resin car.
